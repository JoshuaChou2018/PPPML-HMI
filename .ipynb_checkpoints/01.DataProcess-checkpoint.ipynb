```python
import os
import pandas as pd
import pickle
from collections import Counter

metadata = pickle.load(open('semantic_report_dict.pickle','rb'))
df = pd.read_csv('CT_Scan_Metadata_Complete_35747.csv')

avail_ID = []

for file in os.listdir('rescaled_ct_denoise_float16'):
    if file.endswith('.npz'):
        ID = file.rstrip('.npz')
        if ID in metadata.keys():
            avail_ID.append(ID)
        
df = df.set_index('NoteAcc_DEID')

df = df.loc[avail_ID, :]

infection_to_lung_ratios = []
binary_infection_to_lung_ratios = []
for index in df.index:
    infection_to_lung_ratios.append(metadata[index]['infection_to_lung_ratio'])
    if metadata[index]['infection_to_lung_ratio']>0.1:
        binary_infection_to_lung_ratios.append(1)
    else:
        binary_infection_to_lung_ratios.append(0)
    
df['infection_to_lung_ratio'] = infection_to_lung_ratios
df['binary_infection_to_lung_ratio'] = binary_infection_to_lung_ratios

ManufacturerModelNameCount = Counter(list(df.ManufacturerModelName))
print(ManufacturerModelNameCount)

for key in ManufacturerModelNameCount:
    print(key, len(df[(df.ManufacturerModelName==key)&(df.binary_infection_to_lung_ratio==0)]), len(df[(df.ManufacturerModelName==key)&(df.binary_infection_to_lung_ratio==1)]))
    
split_settings = {'c1': {'SOMATOM Definition Flash':[170,170],'Discovery CT750 HD':[170,170],'LightSpeed VCT':[30,30],'SOMATOM Force':[20,20]},
                  'c2_split1' : {'SOMATOM Definition Flash':[100,100],'Discovery CT750 HD':[100,100]},
                  'c2_split2' : {'SOMATOM Definition Flash':[170,170],'Discovery CT750 HD':[170,170]},
                    'c3_split1' : {'SOMATOM Definition Flash':[170,170],'Discovery CT750 HD':[170,170],'LightSpeed VCT':[30,30]},
                    'c4_split1' : {'SOMATOM Definition Flash':[170,170],'Discovery CT750 HD':[170,170],'LightSpeed VCT':[30,30],'SOMATOM Force':[20,20]}}
                  
dt = df.copy()
dt['client'] = 'NAN'
#exp = 'c2_split'
for exp in split_settings.keys():
    print(exp)
    if exp == 'c1':
        user_id = 0
        for _user_id in range(len(split_settings[exp])):
            user_device = list(split_settings[exp].keys())[_user_id]
            user_sample_numbers = split_settings[exp][user_device]
            neg_samples = list(df[(df.ManufacturerModelName==user_device)&(df.binary_infection_to_lung_ratio==0)].index)
            pos_samples = list(df[(df.ManufacturerModelName==user_device)&(df.binary_infection_to_lung_ratio==1)].index)
            negative_choice = neg_samples[0:user_sample_numbers[0]]
            positve_choice = pos_samples[0:user_sample_numbers[1]]
            user_choice = list(positve_choice) + list(negative_choice)
            dt.loc[user_choice,'client'] = user_id
            print(exp, user_id, user_device, len(user_choice))
    else:
        for user_id in range(len(split_settings[exp])):
            user_device = list(split_settings[exp].keys())[user_id]
            user_sample_numbers = split_settings[exp][user_device]
            neg_samples = list(df[(df.ManufacturerModelName==user_device)&(df.binary_infection_to_lung_ratio==0)].index)
            pos_samples = list(df[(df.ManufacturerModelName==user_device)&(df.binary_infection_to_lung_ratio==1)].index)
            negative_choice = neg_samples[0:user_sample_numbers[0]]
            positve_choice = pos_samples[0:user_sample_numbers[1]]
            user_choice = list(positve_choice) + list(negative_choice)
            dt.loc[user_choice,'client'] = user_id
            print(exp, user_id, user_device, len(user_choice))
    dt.to_csv(f'{exp}.csv',sep=',',index=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from covseg.Tool_Functions import Functions
import os
import pandas as pd
import pickle
from collections import Counter
from multiprocessing import Pool

metadata = pickle.load(open('semantic_report_dict.pickle','rb'))
df = pd.read_csv('CT_Scan_Metadata_Complete_35747.csv')

avail_ID = []

for file in os.listdir('rescaled_ct_denoise_float16'):
    if file.endswith('.npz'):
        ID = file.rstrip('.npz')
        if ID in metadata.keys():
            avail_ID.append(ID)

#sample_id = 'trn00039'
def process_ond_id(sample_id):
    print(sample_id)
    ct_array = np.load(f'rescaled_ct_denoise_float16/{sample_id}.npz')['array']
    lung_mask = np.load(f'RAD_semantics_rescaled_ct/lung_mask/{sample_id}.npz')['array']
    lung_ct_array = ct_array*lung_mask

    (x1, x2), (y1, y2), (z1, z2) = Functions.get_bounding_box(lung_mask)
    new_lung_ct_array = lung_ct_array[x1:x2,y1:y2,z1:z2]
    new_lung_ct_array_resized = np.array([Functions.rescale_2d_array(new_lung_ct_array[:,:,i],(70,70)) for i in range(int(new_lung_ct_array.shape[2]/2)-35, int(new_lung_ct_array.shape[2]/2)+35)])
    print(new_lung_ct_array_resized.shape)

    np.save(f'my_rescaled_70x/{sample_id}.npy',new_lung_ct_array_resized)
    
with Pool(80) as p:
    p.map(process_ond_id, avail_ID)
```
